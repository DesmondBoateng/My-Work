```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#1
x=np.zeros(4)
x[0]=0.2
for i in range(0,3):
    x[i+1]=4*x[i]*(1-x[i])
x
```

```python
#2
sigma=0.001
y=np.zeros(4)
for i in range(0,4):
    y[i]=x[i]+np.random.normal(0,sigma)
plt.plot(y,'b*',markersize=4)
plt.show()
```

```python
#3
def I(x2,y2,y3):
    Ixy=(x2-y2)**2+((4*x2*(1-x2))-y3)**2
    return Ixy
```

```python
np.random.seed(42)
y2=y[1]
y3=y[2]
x2=np.random.uniform(-0.1,1.1,4)
m=I(x2,y2,y3)
```

```python
plt.plot(m,x2)
plt.xlabel('I')
plt.ylabel('x2')
plt.show()
```

```python
#4
sigma=0.9
Y=np.zeros(4)
for i in range(0,4):
    Y[i]=x[i]+np.random.normal(0,sigma)
Y
```

```python
Y2=Y[1]
Y3=Y[2]
x2=np.random.uniform(-0.1,1.1,4)
mnew=I(x2,Y2,Y3)
```

```python
plt.plot(mnew,x2)
plt.plot(m,x2)
plt.xlabel('I')
plt.ylabel('x2')
plt.show()
```
